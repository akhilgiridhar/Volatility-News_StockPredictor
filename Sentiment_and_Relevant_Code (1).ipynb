```python
def categorize_sentiment(sentiment_score):
    """Categorize the sentiment score into various categories."""
    if sentiment_score <= -0.6:
        return 'Very Negative'
    elif -0.6 < sentiment_score <= -0.2:
        return 'Slightly Negative'
    elif -0.2 < sentiment_score < 0.2:
        return 'Neutral'
    elif 0.2 <= sentiment_score < 0.6:
        return 'Slightly Positive'
    else:
        return 'Very Positive'

def categorize_relevance(sentiment_score):
    """Categorize the relevance based on the absolute value of the sentiment score."""
    abs_score = abs(sentiment_score)
    if abs_score >= 0.6:
        return 'Very Relevant'
    elif 0.4 <= abs_score < 0.6:
        return 'Moderately Relevant'
    elif 0.2 <= abs_score < 0.4:
        return 'Slightly Relevant'
    else:
        return 'Not Relevant'

# Apply the categorizations to each row
data['Sentiment_GPT'] = data['Sentiment_Average'].apply(categorize_sentiment)
data['Relevance_GPT'] = data['Sentiment_Average'].apply(categorize_relevance)

# Save the updated dataframe to a new CSV file
updated_file_path = '/mnt/data/Updated_Headlines2022_Sentiment_Analysis.csv'
data.to_csv(updated_file_path, index=False)

updated_file_path
```
